Instruction Tuning: "Teaching the Model to Follow You"

Demonstrates how structured instructions transform LLM outputs

from generic to professional, compliant, and empathetic responses.

## Loan Rejection Emails

```python
# Customer case
customer = {
    "name": "Sarah Johnson",
    "reason": "credit_score", 
    "product": "Personal Loan"
}
basic_prompt = f"Write email to {customer['name']}: loan rejected due to {customer['reason']}"
```

```python
basic_prompt_response = llm.invoke(basic_prompt)
print(basic_prompt_response.content)
```

```python
Instruction_prompt = f"""
You are a FinTech customer officer. Write professional email to {customer['name']} about {customer['product']} rejection.

Guidelines:
- Start with appreciation
- State decision clearly but gently  
- Give general reason (no specific numbers)
- Offer next steps
- Use empathetic, compliant tone
- Professional closing
"""
```

```python
print("\nInstruction Tuning RESPONSE:")

Instruction_prompt_response = llm.invoke(Instruction_prompt)
print(Instruction_prompt_response.content)
```

ADVANCE EXAMPLE FOR INSTRUCTION TUNING PROMPTING FinTech company use case.

```python
mock_customers = [

    {
        "name": "Priya Shah",
        "product_type": "Credit Card",
        "rejection_reason": "high existing credit utilization",
        "email": "priya.shah@example.com",
        "customer_since": "2019",
        "tier": "Standard",
        "location": "Pune, India"
    },

    {
        "name": "Rohit Khanna",
        "product_type": "Business Loan",
        "rejection_reason": "incomplete business documentation",
        "email": "rohit.khanna@example.com",
        "customer_since": "2020",
        "tier": "Enterprise",
        "location": "Bangalore, India"
    },

    {
        "name": "Neha Iyer",
        "product_type": "Home Loan",
        "rejection_reason": "unstable recent income pattern",
        "email": "neha.iyer@example.com",
        "customer_since": "2018",
        "tier": "Premium",
        "location": "Chennai, India"
    },

    {
        "name": "Sandeep Kulkarni",
        "product_type": "Auto Loan",
        "rejection_reason": "policy criteria not met",
        "email": "sandeep.k@example.com",
        "customer_since": "2022",
        "tier": "Standard",
        "location": "Hyderabad, India"
    }
]
```

```python
default_prompt = f"Write an email to {mock_customers[0]['name']} explaining their {mock_customers[0]['product_type']} was rejected due to {mock_customers[0]['rejection_reason']}."
```

```python
default_response = llm.invoke(default_prompt)
print(default_response.content)
```

```python
instructions = f"""
You are a customer relationship officer at a FinTech company.
Write an empathetic, compliant, and professional email to {mock_customers[0]['name']} explaining why their {mock_customers[0]['product_type']} application was not approved.

CUSTOMER DETAILS:
- Customer Name: {mock_customers[0]['name']}
- Product Applied: {mock_customers[0]['product_type']}
- Primary Reason: {mock_customers[0]['rejection_reason']}

FOLLOW THESE STRICT GUIDELINES:

1. TONE & EMPATHY:
   - Begin with genuine appreciation for their application
   - Use empathetic language that acknowledges disappointment
   - Maintain professional yet warm tone throughout
   - Avoid robotic or cold corporate language

2. CLARITY & COMPLIANCE:
   - State the decision clearly in the first paragraph
   - Explain the general reason without technical jargon
   - Never mention specific credit scores, income amounts, or personal details
   - Use compliant language that cannot be interpreted as discriminatory
   - Focus on "policy requirements" rather than personal shortcomings

3. STRUCTURE & CONTENT:
   - Subject line: Clear but not alarming
   - Opening: Thank customer for their trust and application
   - Decision: State rejection clearly but gently
   - Reason: General, policy-based explanation
   - Next steps: Offer constructive path forward
   - Closing: Express hope for future relationship

4. BRAND ALIGNMENT:
   - Use company name: "FinTrust Capital"
   - Include standard compliance disclaimer
   - Offer customer service contact information
   - Maintain brand voice: professional yet approachable

5. FORMATTING:
   - Professional email format with appropriate spacing
   - Natural paragraph breaks for readability
   - Appropriate salutation and closing
   - Include placeholder for customer service details

Avoid any language that could be construed as:
- Discriminatory based on age, gender, ethnicity, etc.
- Making promises or guarantees about future approvals
- Sharing specific underwriting criteria or thresholds
"""
```

```python
response = llm.invoke(instructions)
print(response.content)
```

# Basic Prompting vs. Structured Instruction Prompting specifically for FinTech rejection-email + ticket summarizer scenarios.

| Dimension                         | **Basic Prompting (Raw Prompt)**                   | **Instruction-Based + Role Prompting**        | **Business Impact**                          |
| --------------------------------- | -------------------------------------------------- | --------------------------------------------- | -------------------------------------------- |
| **Output Consistency**            | Highly variable, unpredictable tone                | Stable, repeatable, policy-aligned            | Reduces compliance & brand-voice risk        |
| **Regulatory Compliance**         | Often misses legal wording, may over-share details | Explicit rules prevent violations             | Prevents regulatory breaches & penalties     |
| **Tone & Empathy**                | Robotic or overly casual                           | Warm, empathetic, human-like                  | Better customer trust & reduced complaints   |
| **Actionability**                 | Missing steps or vague summaries                   | Clear conclusions, next steps, and structure  | Shorter resolution times, improved CSAT      |
| **Error Rate**                    | High — model hallucination, missing key details    | Low — instructions constrain behavior         | Prevents misinformation & legal exposure     |
| **Domain Alignment**              | Generic content, not industry-aware                | Deeply aligned to FinTech tone + workflow     | Boosts credibility & professionalism         |
| **Scalability**                   | Inconsistent when scaled across agents             | Uniform responses across thousands of tickets | Operational efficiency & lower training cost |
| **Auditability**                  | Hard to justify or trace decision logic            | Structured outputs → audit-ready              | Crucial for regulated FinTech workflows      |
| **Customer Dissatisfaction Risk** | High — tone mistakes, unclear decisions            | Low — controlled, empathetic narrative        | Fewer escalations, higher NPS                |
