```python
from google_images_download import google_images_download
import os

# Create the downloader object
response = google_images_download.googleimagesdownload()

# âœ… Define your target folder where images will be saved
save_path = os.path.join(os.getcwd(), "downloads", "dogs_images")
os.makedirs(save_path, exist_ok=True)

# âœ… Define search arguments
arguments = {
    "keywords": "Dogs_images",
    "limit": 10,
    "print_urls": True,
    "output_directory": save_path,   # Folder to save images
    "format": "jpg",                 # Optional: Force JPG format
    "no_directory": True             # Avoid creating subfolders per keyword
}

# âœ… Download images
paths = response.download(arguments)

print("\nâœ… Images saved successfully!")
print(f"ðŸ“‚ Saved in: {save_path}")
print(f"ðŸ“¸ Image Paths: {paths}")

```
