```python
from langchain_openai import ChatOpenAI
from langchain.schema import SystemMessage, HumanMessage

# 1. Initialize LLM
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0
)

# 2. System Prompt
system_prompt = """
You are a document comparison assistant.

Compare two documents based on:
1. Overall similarity
2. Key similarities
3. Key differences
4. Missing information
5. Final summary

Return output strictly in JSON format.
"""

# 3. Sample Documents
textA = """
This contract outlines employment terms including salary,
working hours, and termination conditions.
"""

textB = """
This agreement defines employment terms such as salary,
working hours, benefits, and leave policy.
"""

# 4. User Prompt
user_prompt = f"""
TextA:
{textA}

TextB:
{textB}

Compare the documents.
"""

# 5. Run Single Agent
response = llm([
    SystemMessage(content=system_prompt),
    HumanMessage(content=user_prompt)
])

print(response.content)
```
