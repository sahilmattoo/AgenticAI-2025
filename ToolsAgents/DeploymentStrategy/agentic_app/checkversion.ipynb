```python
import sys
import subprocess
from importlib import metadata
import langchain

# cell 0: check langchain version
# Save this as a new file or run the cell in your notebook.


def get_langchain_version():
    # Try attribute on imported package
    try:
        ver = getattr(langchain, "__version__", None)
        if ver:
            return ver
    except Exception:
        pass

    # Try importlib.metadata
    try:
        return metadata.version("langchain")
    except metadata.PackageNotFoundError:
        pass
    except Exception:
        pass

    # Fallback to pip show
    try:
        out = subprocess.check_output([sys.executable, "-m", "pip", "show", "langchain"], text=True)
        for line in out.splitlines():
            if line.lower().startswith("version:"):
                return line.split(":", 1)[1].strip()
    except Exception:
        pass

    return None

version = get_langchain_version()
if version:
    print("langchain version:", version)
else:
    print("langchain is not installed or version could not be determined")
```

```python
# Check versions for langchain (already obtained), langchain-community, and fastapi.
# Reuses imports and `version` variable from previous cell.

def get_package_version(package_name: str):
    # Try importing common module name variants first
    candidates = [package_name, package_name.replace("-", "_")]
    for name in candidates:
        try:
            mod = __import__(name)
            ver = getattr(mod, "__version__", None)
            if ver:
                return ver
        except Exception:
            pass

    # Try importlib.metadata (already imported as metadata)
    try:
        return metadata.version(package_name)
    except Exception:
        pass

    # Fallback to pip show
    try:
        out = subprocess.check_output([sys.executable, "-m", "pip", "show", package_name], text=True)
        for line in out.splitlines():
            if line.lower().startswith("version:"):
                return line.split(":", 1)[1].strip()
    except Exception:
        pass

    return None

langchain_version = version  # keep existing value from the notebook
langchain_community_version = get_package_version("langchain-community")
fastapi_version = get_package_version("fastapi")

print("langchain version:", langchain_version)
print("langchain-community version:", langchain_community_version or "not installed / unknown")
print("fastapi version:", fastapi_version or "not installed / unknown")
```
