```python
context = """You are given excerpts from a High Court judgment and a Supreme Court judgment dealing with 
Article 21 (Right to Life & Liberty). 
Your task is to analyze or answer questions based on this context."""
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

direct_prompt = PromptTemplate(
    input_variables=["context"],
    template="""
You are a legal assistant.

Question:
How does the Supreme Court interpret Article 21 differently from the High Court?

Context:
{context}
"""
)

response = llm.invoke(
    direct_prompt.format(
        context="High Court treats Article 21 as procedural. Supreme Court treats it as intrinsic liberty."
    )
)

print(response.content)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

few_shot_prompt = PromptTemplate(
    input_variables=["hc_reasoning", "sc_reasoning"],
    template="""
You are a legal analyst.

Classify the type of legal reasoning change.

Examples:

High Court Reasoning:
"Article 21 is subject to procedure established by law."

Supreme Court Reasoning:
"Article 21 includes dignity and autonomy."

Classification:
DOCTRINAL

---

High Court Reasoning:
"The restriction is valid under Article 19(2)."

Supreme Court Reasoning:
"Any restriction must satisfy proportionality."

Classification:
SUBSTANTIVE

---

Now classify the following:

High Court Reasoning:
{hc_reasoning}

Supreme Court Reasoning:
{sc_reasoning}

Classification:
"""
)

response = llm.invoke(
    few_shot_prompt.format(
        hc_reasoning="Article 21 is limited by statute.",
        sc_reasoning="Privacy is intrinsic and enforceable."
    )
)

print(response.content)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

cot_prompt = PromptTemplate(
    input_variables=["context"],
    template="""
You are a constitutional law expert.

Analyze the legal evolution step by step.

Step 1: Identify the High Court's interpretation of Article 21.
Step 2: Identify the Supreme Court's interpretation of Article 21.
Step 3: Compare the legal philosophy.
Step 4: Explain the constitutional impact.

Context:
{context}

Provide your reasoning step by step.
"""
)

response = llm.invoke(
    cot_prompt.format(
        context="High Court applied procedural compliance. Supreme Court applied proportionality and dignity."
    )
)

print(response.content)
```
