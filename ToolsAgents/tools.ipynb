## Using LangGraph Prebuilt Agent

```python
from langchain_openai import ChatOpenAI
from langchain.tools import tool
from langgraph.prebuilt import create_react_agent
```

```python
@tool
def mathsHomework(a: int, b: int) -> int:
    """Call this tool if the question is related to math homework."""
    return f"This is Maths Homework Tool" 

@tool
def historyHomework(name: str) -> str:
    """Call this tool if the question is related to history homework."""
    return f"This is History Homework Tool"
```

```python
tools = [mathsHomework, historyHomework]
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
```

```python
agent = create_react_agent(llm, tools)
```

```python
result = agent.invoke({"messages": "Multiply 9 and 7 for my maths homework."})
```

```python
print(result["messages"][-1].content)
```

```python
result = agent.invoke({"messages": "Write a brief note on George Washington for my history homework."})
print(result["messages"][-1].content)
```

#### Explanantions

```python
user_query = "Who was the first president of the United States?"
result = agent.invoke({"messages": user_query})
messages = result["messages"]
tool_called = None
tool_reasoning = None
final_answer = messages[-1].content

# Scan messages to find the *thought* and *action*
for msg in messages:
    if msg.type == "ai":   # LLM thought messages
        if "tool" in msg.additional_kwargs.get("metadata", {}):
            tool_reasoning = msg.content

    if msg.type == "tool":  # Tool execution message
        tool_called = msg.name

if tool_called:
    print(f"ðŸ”§ Tool Selected: {tool_called}")
else:
    print("ðŸ”§ No tool was selected.")

if tool_reasoning:
    print(f"ðŸ¤– Reason for choosing tool:\n{tool_reasoning}")
else:
    print("ðŸ¤– No explicit reasoning provided.")

print(final_answer)
```
