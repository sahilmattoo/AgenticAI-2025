## DEMO: Logging & Tracing Basics with LangSmith

### INTRODUCTION
This demo introduces how to use **LangSmith** for automatic **logging and tracing** of LLM activity.  
Unlike manual file-based logs, LangSmith provides structured observability across prompts, tool calls, reasoning steps, and model outputs — all stored in a web dashboard for visual debugging.

We’ll:
- Configure environment variables for LangSmith.
- Create a simple agent that uses OpenAI via LangChain.
- Observe traces in the LangSmith dashboard, including inputs, outputs, and latency.

````python
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langsmith import traceable  # still used for explicit tracing

# ---------------------------------------------------------
# 1) Environment Setup
# ---------------------------------------------------------
# 1️  Set credentials
import os
os.environ["OPENAI_API_KEY"] = ""
os.environ["LANGCHAIN_API_KEY"] = "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"
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = "LangSmith_Tracing_Demo"

# ---------------------------------------------------------
# 2) Model and Prompt
# ---------------------------------------------------------
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
prompt = ChatPromptTemplate.from_template(
    "You are a helpful assistant. Answer concisely.\n\nQuestion: {question}"
)

# ---------------------------------------------------------
# 3) Traceable Chain Function
# ---------------------------------------------------------
@traceable(run_type="chain")
def ask_agent(question: str):
    chain = prompt | llm
    result = chain.invoke({"question": question})
    return result.content

# ---------------------------------------------------------
# 4) Test Queries
# ---------------------------------------------------------
if __name__ == "__main__":
    queries = [
        "Who discovered penicillin?",
        "Explain the difference between AI and machine learning.",
        "What is the square root of 256?"
    ]

    for q in queries:
        print(f"Query: {q}")
        print("Answer:", ask_agent(q))
        print("-" * 80)
````
