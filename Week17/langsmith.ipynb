```python
import os

os.environ["LANGCHAIN_TRACING_V2"] = "true"
#os.environ["LANGCHAIN_API_KEY"] = "your_langsmith_api_key"
os.environ["LANGCHAIN_PROJECT"] = "Week17_Agent_Evaluation"
#os.environ["OPENAI_API_KEY"] = "your_openai_key"
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o-mini")
response = llm.invoke("Explain agent evaluation in simple terms.")
print(response.content)
```

```python
from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
prompt = ChatPromptTemplate.from_template(
    "Explain the concept of {topic} in simple terms."
)
chain = prompt | llm | StrOutputParser()
result = chain.invoke({"topic": "agent evaluation"})
print(result)
```

```python
def run_agent(query):
    result = chain.invoke({"topic": query})
    return result
```

```python
run_agent("Agent evaluation in AI systems")
```

```python
from langsmith import traceable

@traceable
def run_agent(topic):
    print("Running custom function...")
    
    result = chain.invoke({"topic": topic})
    
    return result
```

```python
run_agent("Observability in AI systems")
```

### Adding Multiple functions for Tracing

```python
@traceable
def retrieve_info(topic):
    return f"Information about {topic}"

@traceable
def generate_answer(info):
    return chain.invoke({"topic": info})

@traceable
def run_agent(topic):
    
    info = retrieve_info(topic)
    
    answer = generate_answer(info)
    
    return answer
```

```python
run_agent("AI agents")
```

## Simple Monitoring Script

```python
from langsmith import Client

client = Client()

runs = client.list_runs(project_name=os.environ["LANGCHAIN_PROJECT"] )

total_latency = 0
total_runs = 0
errors = 0

for run in runs:
    
    total_runs += 1
    
    latency = (run.end_time - run.start_time).total_seconds()
    total_latency += latency
    
    if run.error:
        errors += 1

avg_latency = total_latency / total_runs
error_rate = errors / total_runs

print("Average Latency:", avg_latency)
print("Error Rate:", error_rate)
```
