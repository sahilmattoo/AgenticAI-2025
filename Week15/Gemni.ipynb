```python
from dotenv import load_dotenv
import os

# Load environment variables from .env in the current working directory
load_dotenv()

# Try common names for the Gemini key in .env
gemini_key = (
    os.getenv("GEMINI_API_KEY")
    or os.getenv("GEMINI_KEY")
    or os.getenv("GOOGLE_API_KEY")
)

if not gemini_key:
    raise RuntimeError("Gemini API key not found. Add GEMINI_API_KEY to your .env file.")

# Ensure a key is available under a name your client might expect (won't overwrite existing)
os.environ.setdefault("GENAI_API_KEY", gemini_key)

# Optional: print a masked confirmation
print("Gemini key loaded (masked):", gemini_key[:4] + "..." if gemini_key else None)
```

```python
from google import genai

# use the existing gemini_key variable from the notebook
#client = genai.Client(api_key=gemini_key)
client = genai.Client()

def test_gemini_llm():
    try:
        resp = client.models.generate_content(
            #model="gemini-3-flash-preview",
            model="gemini-2.5-flash-lite",
            contents="Reply with the single word: READY",
            #max_output_tokens=10,
        )
        text = resp.text.strip()
        if text.split() and text.split()[0].upper().startswith("READY"):
            print("Gemini LLM test passed:", text)
        else:
            print("Gemini LLM test failed. Unexpected response:", text)
    except Exception as e:
        print("Gemini LLM test error:", e)

test_gemini_llm()
```
